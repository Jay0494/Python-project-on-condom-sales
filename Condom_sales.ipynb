```python
# Having understood my dataset and requirements for analysis i will drop irrelevant columns
df = df.drop(columns = ['Sex Education Programs (Yes/No)','Male vs Female Purchases (%)','HIV Prevention Awareness (%)', 'Contraceptive Usage Rate (%)'])
```

```python
df.head(6)
```

```python
# get a summary of my datatypes
df.info()
```

```python
# The year column has just the year value which if wouldnt work well as a date soo i used this process to get a more valid date
df['Year'] = pd.to_datetime(df['Year'].astype(str) + '-01-01')
print(df)
```

```python
df.describe()
```

```python
# View my data summary 
df.info()
```

```python
# I want to change some column names 
df =  df.rename(columns = {'Total Sales (Million Units)': 'Total units sold'})
print(df)
```

```python
df.head()
```

```python
df['Market Revenue (Million USD)'] = df['Market Revenue (Million USD)']*1000000
print(df)
```

```python
df.head(7)
```

```python
# to set my python systems to give me actual values rather than scientific annotations
pd.set_option('display.float_format', '{:.6f}'.format)
```

```python
df.head(7)
```

```python
df.describe()
```

```python
correlation = df['Total units sold'].corr(df['Market Revenue (Million USD)'])
print('correlation between sales and revenue:', correlation)
```

```python
plt.figure(figsize = (10, 6))
sns.scatterplot(x=df['Total units sold'], y = df['Market Revenue (Million USD)'], alpha = 0.5)
plt.title('Correlation between Units sold and revenue')
plt.xlabel('Units sold')
plt.ylabel('Market revenue')
plt.show()
```
